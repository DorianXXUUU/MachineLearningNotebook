normalization
适用于分布有明显便捷的情况; 受outlier影响较大
eg:学生分数, 像素点

将所有的数据映射到统一尺度
最值归一化: 把所有数据映射到0-1之间


## 均值方差归一化 standardization
### 把所有数据归一到均值为0方差为1的分布中, 适用于数据分布没有明显的边界, 有可能存在极端数据值

# 数据归一化处理

```python
import numpy as np
import matplotlib.pyplot as plt
```

## 最值归一化 Normalization

```python
x = np.random.randint(0, 100, size=100)
```

```python
x
```

```python
(x - np.min(x)) / (np.max(x)- np.min(x))
```

```python
X = np.random.randint(0, 100, (50, 2))
```

```python
X[:10, :]
```

```python
X = np.array(X, dtype=float)
```

```python
X[:10, :]
```

```python
X[:, 0] = (X[:, 0] - np.min(X[:,0])) / (np.max(X[:,0]) - np.min(X[:,0]))
X[:, 1] = (X[:, 1] - np.min(X[:,1])) / (np.max(X[:,1]) - np.min(X[:,1]))
```

```python
X[:10, :]
```

```python
plt.scatter(X[:, 0], X[:,1])
plt.show()
```

```python
np.mean(X[:,0])
```

```python
np.mean(X[:,1])
```

```python
np.std(X[:, 0])
```

## 均值方差归一化 Standardization

```python
X2 = np.random.randint(0, 100, (50, 2))
```

```python
X2 = np.array(X2, dtype=float)
```

```python
X2[:10,:]
```

```python
X2[:,0] = (X2[:,0] - np.mean(X2[:,0])) / np.std(X2[:, 0])
```

```python
X2[:,1] = (X2[:,1] - np.mean(X2[:,1])) / np.std(X2[:, 1])
```

```python
X2[:10, :]
```

```python
plt.scatter(X2[:,0], X2[:,1])
plt.show()
```

```python
np.mean(X2[:,0])
```

```python
np.std(X2[:,0])
```
